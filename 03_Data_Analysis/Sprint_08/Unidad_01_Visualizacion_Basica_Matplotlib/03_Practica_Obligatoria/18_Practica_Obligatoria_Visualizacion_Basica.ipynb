Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro.  

Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)  

2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand ITC", tamao 18, color azul acero o "steelblue")
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
crowded_cities = df.sort_values("population_total", ascending=False).head(5)
crowded_cities.head(5)
```

```python
cities = crowded_cities["city"]
population = crowded_cities["population_total"]
```

```python
plt.figure(figsize=(8,5))
plt.bar(cities, population, color = ["blue", "grey", "red", "olive", "brown"], alpha = 0.5)
plt.title("Ciudades ms pobladas de California")
plt.ylabel("Habitantes (miles)")

for i, valor in enumerate(population):
    plt.text(i, valor + 50000, str(valor), ha = "center")
```

```python
df["area_total_km2"].head(5)
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["area_total_km2"], bins = 50, color = "steelblue", alpha = 0.5)
plt.xlabel("Superficie (km)", fontname = "Bradley Hand", fontsize = 18)
plt.ylabel("Nmero de Ciudades", fontname = "Bradley Hand", fontsize = 18);
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["area_water_km2"], df["population_total"], color = "green", alpha = 0.5, s = 20)
plt.title("Poblacin por kilmetro cuadrado de agua")
plt.xlabel("rea de agua en la ciudad (km)")
plt.ylabel("Poblacin (miles habitantes)")
plt.text(60, 3500000, "Outliers", fontsize = 8)
plt.annotate("", xy = (85, 3825000), xytext = (71, 3531250), arrowprops = dict(arrowstyle = "->", color = "black", lw = 1))
plt.annotate("", xy = (120, 1450000), xytext = (71, 3531250), arrowprops = dict(arrowstyle = "->", color = "black", lw = 1))
```

```python
elevation_m = df["elevation_m"].dropna()
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(elevation_m, vert=False, patch_artist=True, boxprops=dict(facecolor='red'), medianprops=dict(color='black'))
plt.title("Diagrama de caja de la Altitud de las ciudades de California")
plt.xlabel("Altitud (m)")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(24,15))
axes[0,0].bar(cities, population, color = ["blue", "grey", "red", "olive", "brown"], alpha = 0.5)
axes[0,0].set_title("Ciudades ms pobladas de California")
axes[0,0].set_ylabel("Habitantes (miles)")

for i, valor in enumerate(population):
    axes[0,0].text(i, valor + 50000, str(valor), ha = "center")

axes[0,1].hist(df["area_total_km2"], bins = 50, color = "steelblue", alpha = 0.5)
axes[0,1].set_xlabel("Superficie (km)", fontname = "Bradley Hand", fontsize = 18)
axes[0,1].set_ylabel("Nmero de Ciudades", fontname = "Bradley Hand", fontsize = 18)

axes[1,0].scatter(df["area_water_km2"], df["population_total"], color = "green", alpha = 0.5, s = 20)
axes[1,0].set_title("Poblacin por kilmetro cuadrado de agua")
axes[1,0].set_xlabel("rea de agua en la ciudad (km)")
axes[1,0].set_ylabel("Poblacin (miles habitantes)")
axes[1,0].text(60, 3500000, "Outliers", fontsize = 8)
axes[1,0].annotate("", xy = (85, 3825000), xytext = (71, 3531250), arrowprops = dict(arrowstyle = "->", color = "black", lw = 1))
axes[1,0].annotate("", xy = (120, 1450000), xytext = (71, 3531250), arrowprops = dict(arrowstyle = "->", color = "black", lw = 1))

axes[1,1].boxplot(elevation_m, vert=False, patch_artist=True, boxprops=dict(facecolor='red'), medianprops=dict(color='black'))
axes[1,1].set_title("Diagrama de caja de la Altitud de las ciudades de California")
axes[1,1].set_xlabel("Altitud (m)")

fig.savefig("practica_obligatoria_2.png", dpi=300)
```
